# Student Performance Prediction

## Data Understanding

### Problem Satement

- Educational institutions often struggle to identify students who may perform poorly in exams before the results are announced. Factors such as study habits, attendance, parental education, socioeconomic background, and previous academic performance can influence a student's final score. By analyzing these factors, we can build a machine learning model that predicts student performance, helping educators identify at-risk students early and provide targeted support.

### Business Problem
Schools and colleges want to answer questions like:

- Which students are likely to score low?
- Which factors affect student performance the most?
- Can we identify struggling students before exams?
- How can teachers provide early intervention?

### Machine Learning Proble

Build a supervised machine learning regression model that predicts a student's exam score based on the available features in the dataset.

### Objective 

- Understand the dataset.
- Perform EDA.
- Analyze relationships between features and G3.
- Clean and preprocess data.
- Encode categorical variables.
- Train multiple regression models.
- Compare their performance.
- Select the best model.
- Identify the most important factors affecting grades.
- Save the trained model.
